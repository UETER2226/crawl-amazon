```python
from PIL import Image
from io import BytesIO
import csv
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
with open("urlamazon.txt", "r") as file:
    content = file.read()

# Tách các dòng thành danh sách các mùa giải
url_list = content.split("\n")

# Loại bỏ khoảng trắng và chuyển thành chữ thường
url_list = [url.strip().lower() for url in url_list if url.strip()]

# In danh sách mùa giải
print(url_list)
```

```python
# Replace 'URL' with the actual URL of the website you want to scrape
urls = []
for url in url_list:
  HEADERS = ({'User-Agent':'Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/44.0.2403.157 Safari/537.36','Accept-Language': 'en-US, en;q=0.5'})
  # Send a GET request to the website
  response = requests.get(url,headers=HEADERS)


  if response.status_code == 200:
      soup = BeautifulSoup(response.content, 'html.parser')
      hrefs = soup.find_all('a', class_='a-link-normal s-underline-text s-underline-link-text s-link-style a-text-normal')
      for href in hrefs:
          tmp = href.get('href')
          urls.append(tmp)
```

```python
def crawl_data(URL):
	with requests.session() as sr:
		sr.headers.update({'User-Agent': 'Mozilla/5.0 (X11; Ubuntu; Linux x86_64; rv:78.0) Gecko/20100101 Firefox/78.0'})
		partUrl = sr.get(URL)
		soup = BeautifulSoup(partUrl.content,'html.parser')

		# retrieving product title
		try:
			# Outer Tag Object
			title = soup.find("span",
							attrs={"id": 'productTitle'})
			# Inner NavigableString Object
			title_value = title.string
			# Title as a string value
			title_string = title_value.strip().replace(',', '')
		except AttributeError:
			title_string = "NA"


		# retrieving price
		try:
			price = next((el.text for el in soup.find_all('span', {'aria-hidden': 'true'}) if el.text.startswith("$")), "NA")
		except AttributeError:
			price = "NA"



		# retrieving product rating
		try:
			rating = soup.find("i", attrs={'class': 'a-icon a-icon-star a-star-4-5'}).string.strip().replace(',', '')

		except AttributeError:

			try:
				rating = soup.find(
					"span", attrs={'class': 'a-icon-alt'}).string.strip().replace(',', '')
			except:
				rating = "NA"


		try:
			review_count = soup.find("span", attrs={'id': 'acrCustomerReviewText'}).string.strip().replace(',', '')

		except AttributeError:
			review_count = "NA"


		# print availablility status
		try:
			available = soup.find("div", attrs={'id': 'availability'})
			available = available.find("span").string.strip().replace(',', '')

		except AttributeError:
			available = "NA"

		# product detail
		try:
			detail = soup.find_all("div", class_ = "a-fixed-left-grid-inner", style = "padding-left:140px")
			span_tags = []
			for div in detail:
				spans = div.find_all("span", class_="a-color-base")
				for span in spans:
					span_tags.append(span.get_text())
			dic = {}
			for i in range(0,len(span_tags),2):
				dic[span_tags[i]] = span_tags[i+1]
		except AttributeError:
			dic = "NA"


		# product description
		try:
			a = soup.find_all("span",class_= "a-list-item a-size-base a-color-base")
			product_description =""
			for div in a:
				tmp = div.get_text()
				product_description += tmp
				product_description += ". "
		except AttributeError:
			product_description = "NA"

	# url product image
		try:
			img_tag = soup.find("img", id="landingImage")
			if img_tag is not None:
					image_url = img_tag["src"]
			else:
					image_url = "NA"
		except AttributeError:
			image_url = "NA"

		return title_string,price,rating,review_count,dic,product_description,image_url


#title_string,price,rating,review_count,available,dic,product_description,image_url = main(urls[1])
```

```python
columns= ["Title","Price","Rating","Review Count","Detail","Description","Image_url","Department"]
data = []
```

```python
for url in urls:
  title_string,price,rating,review_count,dic,product_description,image_url = crawl_data(url)
  row_data = [title_string,price,rating,review_count,dic,product_description,image_url,"Clothig for girl"]
  data.append(row_data)
```

```python
products = pd.DataFrame(data, columns=columns)
```
